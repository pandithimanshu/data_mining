```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
from sklearn.preprocessing import StandardScaler
numerical_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)']
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

df
```

```python
#standardization: Handles Different Scales
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df
```

```python
#transformation : Handles Zero Values Gracefully, Reduces Skewness
for col in numerical_cols:
    df[col] = np.log(df[col])
df.columns
```

```python
df_agg = df.groupby('target').agg({
    'sepal length (cm)': ['mean', 'min', 'max'],
    'sepal width (cm)': ['mean', 'min', 'max'],
    'petal length (cm)': ['mean', 'min', 'max'],
    'petal width (cm)': ['mean', 'min', 'max']
})

print("\nMultiple Aggregations (Mean, Min, Max):")
print(df_agg)
```

```python
print("inf values:", np.isinf(df.select_dtypes(include=[np.number]).to_numpy()).sum())


numeric_data = df.select_dtypes(include=[np.number])

mask = ~np.isinf(numeric_data.to_numpy()).any(axis=1)

df = df[mask]

print("Remaining inf values:", np.isinf(df.select_dtypes(include=[np.number]).to_numpy()).sum())
```

```python
df['Sepal.Length_bins'] = pd.cut(df['sepal length (cm)'], bins=3, labels=['Short', 'Medium', 'Long'])
print("\nSepal Length Bins:")
print(df[['sepal length (cm)', 'Sepal.Length_bins']].head())
```
