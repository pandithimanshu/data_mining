```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print("Performance for each K:")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(silhouette)
    print(f"K={k} | Inertia={kmeans.inertia_:.2f} | Silhouette Score={silhouette:.4f}")
```

```python
# Step 5: Final clustering using the optimal k = 3
k_optimal = 3
kmeans_final = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
kmeans_final.fit(X_scaled)
labels = kmeans_final.labels_
centroids = kmeans_final.cluster_centers_

# Step 6: Display clustering result
print("\nCluster labels for first 10 samples:", labels[:10])
print("Shape of centroid array:", centroids.shape)
```
