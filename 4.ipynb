```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
pd.DataFrame(data = iris.data,columns =iris.feature_names)
```

```python
x = iris.data
y = iris.target
```

```python
x.shape
```

```python
y.shape
```

Hold out


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

k fold

```python
from sklearn.model_selection import KFold
```

```python
x = [0,1,2,3,4,5,6,7,8]
```

```python
kf = KFold(n_splits=2)
```

```python
kf
```

```python
for train,test in kf.split(x):
  print(train,test)
```

```python
kf2 = KFold(n_splits=5)
```

```python
for train,test in kf2.split(x):
  print(train,test)
```

```python
kf3 = KFold(n_splits=9)
for train,test in kf3.split(x):
  print(train,test)
```

k nearest neighbour classification using holdout method:

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
prediction_knn = knn.predict(x_test)
```

classification metrics

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy on the test data")
print(accuracy_score(y_test,prediction_knn))

print("classification report on the test data")
print(classification_report(y_test,prediction_knn))
```
